```python
import pandas as pd
```

```python
dfs = []
df7 = pd.read_csv("net7df.csv")
dfs.append(df7)
for i in range(14, 57, 7):
    dfi = pd.read_csv("interval" + str(i) + "df.csv")
    dfs.append(dfi)
```

```python
union = set(dfs[0].ID.to_numpy())
for i in range(1, 8):
    union = union & set(dfs[i].ID.to_numpy())
    print(len(union))
```

```python
union1 = set(dfs[0].ID.to_numpy()) | set(dfs[1].ID.to_numpy()) |set(dfs[2].ID.to_numpy()) | set(dfs[3].ID.to_numpy())
union2 = set(dfs[4].ID.to_numpy()) | set(dfs[5].ID.to_numpy()) |set(dfs[6].ID.to_numpy()) | set(dfs[7].ID.to_numpy())
union = union1 & union2
print(len(union))
```

```python
for i in range(0, 8):
    data = dfs[i].loc[dfs[i]['ID'].isin(union)]
    data = data.reset_index(drop = True)
    #data.to_csv("Moodable" + str(i) + ".csv", index = False)
```
